```python
def dimension(g):
    g = np.exp(-g)
    eig = np.clip(np.linalg.eigvalsh(g), 10e-6, None)
    eig = eig / np.sum(np.abs(eig))
    return np.exp(np.sum(- eig * np.log(eig)))

form = 'png'
```

```python
names = ['SPC','SPC/E','SPC/Eb','TIP3P','TIP3P-FB',
         'OPC3','TIP4P','TIP4P-EW','TIP4P/2005','TIP4P-ICE',
         'TIP4P/ε','TIP4P-FB','OPC','TIP5P','TIP5P-EW',
         'TIP5P/2018']

spectra = []

# 3-sites models
spectra.append(np.mean(np.load('references_water/spc_soap.npz')['arr_0'],axis=0))
spectra.append(np.mean(np.load('references_water/spce_soap.npz')['arr_0'],axis=0))
spectra.append(np.mean(np.load('references_water/spceb_soap.npz')['arr_0'],axis=0))
spectra.append(np.mean(np.load('references_water/tip3p_soap.npz')['arr_0'],axis=0))
spectra.append(np.mean(np.load('references_water/tip3pfb_soap.npz')['arr_0'],axis=0))
spectra.append(np.mean(np.load('references_water/opc3_soap.npz')['arr_0'],axis=0))

# 4-sites models
spectra.append(np.mean(np.load('references_water/tip4p_soap.npz')['arr_0'],axis=0))
spectra.append(np.mean(np.load('references_water/tip4pew_soap.npz')['arr_0'],axis=0))
spectra.append(np.mean(np.load('references_water/tip4p2005_soap.npz')['arr_0'],axis=0))
spectra.append(np.mean(np.load('references_water/tip4pice_soap.npz')['arr_0'],axis=0))
spectra.append(np.mean(np.load('references_water/tip4peps_soap.npz')['arr_0'],axis=0))
spectra.append(np.mean(np.load('references_water/tip4pfb_soap.npz')['arr_0'],axis=0))
spectra.append(np.mean(np.load('references_water/opc_soap.npz')['arr_0'],axis=0))

# 5-sites models
spectra.append(np.mean(np.load('references_water/tip5p_soap.npz')['arr_0'],axis=0))
spectra.append(np.mean(np.load('references_water/tip5pew_soap.npz')['arr_0'],axis=0))
spectra.append(np.mean(np.load('references_water/tip5p2018_soap.npz')['arr_0'],axis=0))

for i in range(len(spectra)):
    spectra[i] = spectra[i].astype('float64')



distmat = np.zeros((len(spectra),len(spectra)))

for i in range(len(spectra)):
    for j in range(i+1,len(spectra)):
        distmat[i,j] = DistanceSoap(spectra[i],spectra[j])
        distmat[j,i] = DistanceSoap(spectra[j],spectra[i])
        
#print(distmat)
```

```python
ff = names #[mapping_num[i] for i in range(len(x))]

names = ['SPC','SPC/E','SPC/Eb','TIP3P','TIP3P-FB',
         'OPC3','TIP4P','TIP4P-EW','TIP4P/2005','TIP4P-ICE',
         'TIP4P/ε','TIP4P-FB','OPC','TIP5P','TIP5P-EW',
         'TIP5P/2018']
color=['#CCCCCC','#999999','#666666','#880015','#00205b',
       '#b97a57','#ed1c24','#ff7f27','#fff200','#efe4b0',
       '#22b14c','#ffc90e','#a349a4','#99d9ea','#00a2e8',
       '#3f48cc']

def get_colors(color_list):
    mapping_color = {
        "SPC": "#CCCCCC",
        "SPC/E": "#999999",
        "SPC/Eb": "#666666",
        "TIP3P": "#880015",
        "TIP3P-FB": "#00205b",
        "OPC3": "#b97a57", 
        "TIP4P": "#ed1c24",
        "TIP4P-EW": "#ff7f27",
        "TIP4P/2005": "#fff200",
        "TIP4P-ICE": "#efe4b0",
        "TIP4P/ε": "#22b14c",
        "TIP4P-FB": "#ffc90e",
        "OPC" : "#a349a4",
        "TIP5P": "#99d9ea",
        "TIP5P-EW": "#00a2e8",
        "TIP5P/2018": "#3f48cc",
    }
    return [mapping_color[i] for i in color_list]
```

```python
d = distmat # np.load('../dscribe_4beads_1us/distance_avg_30_ang.npy')
linkage_matrix = linkage(d, "average", optimal_ordering=True)
u = np.array(dendrogram(linkage_matrix, distance_sort=True, orientation='right', ax=None)['leaves'])
u = u[::-1]
v = d[u,:][:,u]
```

```python
ffs = [names[i] for i in u]
df = pd.DataFrame(v, index=ffs, columns=ffs)

center = df.values[np.tril_indices(df.shape[0], k=1)].mean()

g = sns.clustermap(df, center=center, cmap="Blues", method="average",
                   row_colors=get_colors(df.index),
                   dendrogram_ratio=(.4, .4),
                   cbar_pos=(1.02, .15, .03, .45),
                   linewidths=.75, figsize=(8, 8))

g.ax_col_dendrogram.remove()
_ = g.ax_heatmap.set_xticks([])
g.savefig('distmat_4beads_avg_30_ang_referee.{}'.format(form), bbox_inches='tight')
df.to_csv('distmat_4beads_avg_30_ang_referee.csv')
```

```python
from sklearn import manifold

names = ['SPC','SPC/E','SPC/Eb','TIP3P','TIP3P-FB',
         'OPC3','TIP4P','TIP4P-EW','TIP4P/2005','TIP4P-ICE',
         'TIP4P/ε','TIP4P-FB','OPC','TIP5P','TIP5P-EW',
         'TIP5P/2018']
color=['#CCCCCC','#999999','#666666','#880015','#b97a57',
       '#00205b','#ed1c24','#ff7f27','#fff200','#efe4b0',
       '#22b14c','#ffc90e','#a349a4','#99d9ea','#00a2e8',
       '#3f48cc']


markers = ['^', '^', '^', '^', '^',
           '^', 's', 's', 's', 's', 
           's', 's', 's', 'p', 'p',
           'p']
```

```python
d = distmat #np.load('../dscribe_4beads_1us/distance_avg_30_ang.npy')
#d = d[:, np.array(index)][np.array(index), :]
d = (d - np.amin(d)) / (np.amax(d) - np.amin(d))

mds = manifold.MDS(n_components=2, dissimilarity="euclidean", random_state=7, max_iter=100000)
results = mds.fit(d)

coords = results.embedding_

fig, ax = plt.subplots(figsize=(8, 8), dpi=200)

xoff = np.array([0.05] * len(names))

yoff = np.array([-0.05] * len(names))

coords[:,0] -= 0.1 
for i in range(coords.shape[0]):
    ax.scatter(coords[i, 0],coords[i,1], s=600, edgecolor='k', marker=markers[i], label=names[i],c=color[i] )
    #ax.annotate(names[i], (coords[i,0] + xoff[i], coords[i,1] + yoff[i] ))
    
# plt.legend(bbox_to_anchor=(0., 1.1, 1., .102), loc='lower left',
#            ncol=2, mode="expand", borderaxespad=0.)
# ax.set_title("Force Fields distribution " + "Å")

_ = ax.set_xlim((-1.5, 1.))
_ = ax.set_ylim((-0.8, 1.6))

_ = plt.xticks([], [])
_ = plt.yticks([], [])
fig.savefig('ffdistr_4beads_avg_30_ang_referee.png')
fig.savefig('ffdistr_4beads_avg_30_ang_referee.pdf')
```
